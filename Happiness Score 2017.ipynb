```python
from sklearn import preprocessing
l=preprocessing.LabelEncoder()
df['Country']=l.fit_transform(df['Country'])
df['Country'].unique()
```

```python
X=df[["Country","Whisker.high","Whisker.low","Economy..GDP.per.Capita.","Family","Health..Life.Expectancy.","Freedom","Generosity","Trust..Government.Corruption.","Dystopia.Residual"]]
Y=df["Happiness.Score"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,Y_train)
```

```python
pr=lm.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test,pr)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
```

```python
knn.fit(X_train,Y_train)
```

```python
pr1=knn.predict(X_test)
```

```python
r2_score(Y_test,pr)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfc=RandomForestRegressor()
```

```python
rfc.fit(X_train,Y_train)
```

```python
pr1=rfc.predict(X_test)
```

```python
r2_score(Y_test,pr1)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
```

```python
dtr.fit(X_train,Y_train)
```

```python
pr2=dtr.predict(X_test)
```

```python
r2_score(Y_test,pr2)
```
